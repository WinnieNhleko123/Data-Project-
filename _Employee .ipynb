```python
import pandas as pd # importing the csv file.
import matplotlib.pyplot as plt # common library used for creating a wide variety of static, animated, and interactive plots and visualizations.
```

```python
df = pd.read_csv("employees.csv") # displaying the csv file.
```

```python
df # showing the csv file.
```

```python
dg=df.groupby("JOB_ID") # grouping by JOB_ID.
```

```python
dg.first() # returns the first row of the DataFrame
```

```python
r=df.groupby("DEPARTMENT_ID").agg({"SALARY":"mean"}) # calculating the average salary for each department.
```

```python
r.round(2) # rounding off the output to the nearest 2 decimal place.
```

```python
df.nlargest(5,"SALARY") # dispalying the top 3 highest paid employees.
```

```python
df.groupby('DEPARTMENT_ID')['SALARY'].mean().plot(kind='bar') # used to create a bar plot of the average salary by department
plt.xlabel("DEPARTMENT_ID PER DEPARTMENT") # Add axis labels.
plt.ylabel("SALARY PER MONTH") # Add axis labels
plt.title("Average Salary By Department") # Add a title.
plt.grid(True, which='both', linestyle='--', linewidth=0.9) # add grid for better readability
plt.tight_layout() # adjust layout 
plt.show() # Show the plot.
```
